### 2. Collaborative Filtering Based  
To predict rating Rxi that user 'x' will give to item 'i', we need to have:  
1. ratings matrix
2. k nearest neighbours(most similar user/item)

Predict Rxi using formula:  

**Rxi = (Σy Sxy*Ryi)/(Σy Sxy)**  

**Sab = cos(ra,rb)**  

where **centered cosine** is implemented i.e.  
normalize ratings by subtracting each value with its corresponding row mean

### A. USER-BASED  

Find set of k **users** most similar to user 'x' and who have also rated item 'i'

#### KNN

```python
# if you specify algorithm="brute" the algorithm will calculate all distances. Otherwise it may resort to smart heuristics (such as KD-trees)
def kSimilarUsers(userx_id,ratings_matrix,metric,k):
    similarities=[]
    indices=[]
    model_knn = NearestNeighbors(metric = metric,algorithm="brute") 
    model_knn.fit(ratings_matrix)
    userx_index = ratings_matrix.index.get_loc(userx_id) #taking index of that user
    # getting the ratings vector (rx) of that user i.e. take row userx, every column
    # taking values(ratings) from rx vector and passing it to model
    distances, indices = model_knn.kneighbors(ratings_matrix.iloc[userx_index, :].values.reshape(1, -1), n_neighbors = k+1)
    similarities = 1-distances
    return similarities,indices
```

```python
kSimilarUsers(4385,ratings_matrix,'cosine',10)
```

So, we need to predict rating rxi for an item 'i' by a user 'x'  
i.e. **given userx_id and itemi_id -> rxi=?**  


```python
def predictRating_userBased(userx_id,itemi_id,ratings_matrix,metric='cosine',k=10):
    
    similarities, indices=kSimilarUsers(userx_id, ratings_matrix,metric, k)
    indices = indices.flatten()
    similarities = similarities.flatten()
    n = len(indices)

    userx_index = ratings_matrix.index.get_loc(userx_id)
    itemi_index = ratings_matrix.columns.get_loc(itemi_id)
    
    rxi = 0
    for y in range(0,n): # for every n similar users
        #if indices[y] == userx_index:
            #continue
        Sxy = similarities[y] #sim(userx, current_similar_user_y)
        Ryi = ratings_matrix.iloc[indices[y],itemi_index] # rating vector of user y for item i
        rxi += Sxy*Ryi
    Sxy_sum = np.sum(similarities)-1  # -1 to remove similarity of user with itself
    rxi = rxi/Sxy_sum
    prediction = int(round(rxi))
    
    #in case of very sparse datasets, using correlation metric for collaborative based approach may give negative ratings
    #which are handled here as below //code has been validated without the code snippet below, below snippet is to avoid negative
    #predictions which might arise in case of very sparse datasets when using correlation metric
    if prediction <= 0:
        prediction = 1   
    elif prediction >10:
        prediction = 10
    
    return prediction
```

### A. ITEM-BASED  

```python
def kSimilarItems(itemi_id,ratings_matrix,metric,k):
    similarities=[]
    indices=[]
    ratings_matrix=ratings_matrix.T
    model_knn = NearestNeighbors(metric = metric,algorithm="brute") 
    model_knn.fit(ratings_matrix)
    itemi_index = ratings_matrix.index.get_loc(itemi_id) #taking index of that item
    #print(itemi_index)
    # getting the ratings vector (rx) of that item i.e. take column itemi, every row?
    # taking values(ratings) from rx vector and passing it to model
    #print(ratings_matrix.iloc[itemi_index, :])
    distances, indices = model_knn.kneighbors(ratings_matrix.iloc[itemi_index, :].values.reshape(1, -1), n_neighbors = k+1)
    similarities = 1-distances
    return similarities,indices
```

```python
kSimilarItems('0375707972',ratings_matrix,'cosine',10)
```

```python
def predictRating_itemBased(userx_id,itemi_id,ratings_matrix,metric='cosine',k=10):
    
    similarities, indices=kSimilarItems(itemi_id, ratings_matrix,metric, k)
    indices = indices.flatten()
    similarities = similarities.flatten()
    n = len(indices)

    userx_index = ratings_matrix.index.get_loc(userx_id)
    itemi_index = ratings_matrix.columns.get_loc(itemi_id)
    
    rxi = 0
    for j in range(0,n): # for every n similar item
        #if indices[j] == itemi_index:
            #continue
        Sij = similarities[j] #sim(itemi, current_similar_item_j)
        #print(Sij,itemi_index,indices[j])
        #print(userx_index,indices[j])
        #print(ratings_matrix.iloc[userx_index,indices[j]])
        Rxj = ratings_matrix.iloc[userx_index,indices[j]] # rating vector of item j by user x
        #print(Rxj)
        rxi += Sij*Rxj
    Sxy_sum = np.sum(similarities)-1  # -1 to remove similarity of user with itself
    rxi = rxi/Sxy_sum
    prediction = int(round(rxi))
    
    #in case of very sparse datasets, using correlation metric for collaborative based approach may give negative ratings
    #which are handled here as below //code has been validated without the code snippet below, below snippet is to avoid negative
    #predictions which might arise in case of very sparse datasets when using correlation metric
    if prediction <= 0:
        prediction = 1   
    elif prediction >10:
        prediction = 10
    
    return prediction
```

```python
predictRating_itemBased(4017,'0375707972',ratings_matrix)
```

```python
def recommendItem(user_id, ratings, basis,metric='cosine'):
    prediction = []
    if basis=="item":
        for i in range(ratings.shape[1]):
            if (ratings[str(ratings.columns[i])][user_id] !=0): #not rated already
                prediction.append(predictRating_itemBased(user_id, str(ratings.columns[i]) ,ratings, metric))
            else:  
                prediction.append(-1) #for already rated items
    elif basis == "user":
        for i in range(ratings.shape[1]):
            if (ratings[str(ratings.columns[i])][user_id] !=0): #not rated already
                prediction.append(predictRating_userBased(user_id, str(ratings.columns[i]) ,ratings, metric))
            else:  
                prediction.append(-1) #for already rated items
    prediction = pd.Series(prediction)
    prediction = prediction.sort_values(ascending=False)
    recommended = prediction[:10]
    for i in range(len(recommended)):
        print("{0}. {1}".format(i+1,books.bookTitle[recommended.index[i]]))
```
